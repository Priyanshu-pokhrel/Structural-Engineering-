```python
#importing different libraries
import math
import numpy as np                 #for working with arrays
import matplotlib.pyplot as plt    #plotting functunalities
```

```python
#setting default values
pointloads=np.array([[]])
pointmoments=np.array([[]])
distributedloads=np.array([[]])
linearloads=np.array([[]])
```

```python
#input span and force data
#all lengths are in meter
span=17 
A=3          #distance of support from left
B=13         #distance of second support from left

#force data
#pointloads=np.array([[6,0,-90]])          #pointloads(location,xmagnitude,ymagnitude)
#pointmoments=np.array([[17,50]])          #pointmoments(Location,magnitude) 
#distributedloads=np.array([[8,17,-10]])    #distributedloads(start,end,magnitude)
linearloads=np.array([[8,17,0,-10]])        #linearloads(start loacation,end location,startmag,endmag)

print(pointmoments[0])
```

```python
#range of x co-ordinates
X=np.linspace(0,span,1000)
npl=len(pointloads[0])          #test for pointloads to consider
npm=len(pointmoments[0])        #test for pointmoments to comsider
nUDL=len(distributedloads[0])   #test for UDL to consider
nUVL=len(linearloads[0])        #test for UVL to consider

reactions=np.array([0.0,0,0])     #initalizing the reaction values(va,ha,vb)
shear1=np.empty([0,len(X)])
moment=np.empty([0,len(X)])

print(nUVL)
```

# Reaction calculation

**define a function to calculate reactions of different point loads**

```python
def reactions_PL(n):
    xp=pointloads[n,0]      #location of pointload
    fx=pointloads[n,1]      #horizontal magnitude of a pointload
    fy=pointloads[n,2]      #vertical magnitude of a pointload
    
    la_p=A-xp               #leverarm of pointload
    mp=fy*la_p              #moment at support due to point load
    la_vb=B-A    
    
    vb=mp/la_vb
    va=-fy-vb
    ha=-fx
    
    return va,ha,vb
    
    
```

**define a function to calculate reactions of different point moments**

```python
def reactions_PM(n):
    xm=pointmoments[n,0]       #location of pointmoment
    m=pointmoments[n,1]        #magnitude of pointmoment
    
    la_vb=B-A
    vb=m/(la_vb)
    va=-vb
    
    return va,vb
```

**define a function to calculate reaction due to UDLs**

```python
def reactions_UDL(n):
    x_start=distributedloads[n,0]     #starting position of UDL
    x_end=distributedloads[n,1]       #ending position of UDL
    fy=distributedloads[n,2]         #magnitude of UDl in kn/m
    
    fy_res=fy*(x_end-x_start)          #resultant of UDL
    x_res=x_start+(x_end-x_start)/2    #location of resultant load
    
    la_p=A-x_res                #leverarm of resultant load
    mp=fy_res*la_p              #moment at support due to resultant load
    la_vb=B-A    
    
    vb=mp/la_vb
    va=-fy_res-vb
    
    
    return va,vb
```

**define a function to calculate reaction due to UVLs**

```python
def reactions_UVL(n):
    x_start=linearloads[n,0]          #starting position of UVL
    x_end=linearloads[n,1]            #ending position of UVL
    fy_start=linearloads[n,2]         #start magnitude of UVl in kn/m
    fy_end=linearloads[n,3]           #end magnitude of UVL in kn/m
    
    if(fy_end==0):
        fy_res=0.5*fy_start*(x_end-x_start)
        x_res=x_start+(x_end-x_start)/3
    if(fy_start==0):
        fy_res=0.5*fy_end*(x_end-x_start)
        x_res=x_start+(x_end-x_start)*(2/3)
    
    la_p=A-x_res                #leverarm of resultant load
    mp=fy_res*la_p              #moment at support due to resultant load
    la_vb=B-A    
    
    vb=mp/la_vb
    va=-fy_res-vb
    
    
    return va,vb
```

**cycle through all point loads and determine reactions**

```python
PL_record=np.empty([0,3])
if npl>0:
    for n, p in enumerate(pointloads):
        va,ha,vb=reactions_PL(n)
        PL_record=np.append(PL_record,[np.array([va,ha,vb])],axis=0)
         #add reactions to the record
        reactions[0]=reactions[0]+va
        reactions[1]=reactions[1]+ha
        reactions[2]=reactions[2]+vb

        
       
```

**cycle through all point moments and determine reactions**

```python
PM_record=np.empty([0,2])
if(npm)>0:
    for n,p in enumerate(pointmoments):
        va,vb=reactions_PM(n)
        PM_record=np.append(PM_record,[np.array([va,vb])],axis=0)
        
         #add reactions to the record
        reactions[0]=reactions[0]+va
        reactions[2]=reactions[2]+vb

        
        
```

**cycle through all UDLs and determine reactions**

```python
UDL_record=np.empty([0,2])
if(nUDL)>0:
    for n,p in enumerate(distributedloads):
        va,vb=reactions_UDL(n)
        UDL_record=np.append(UDL_record,[np.array([va,vb])],axis=0)
        
         #add reactions to the record
        reactions[0]=reactions[0]+va
        reactions[2]=reactions[2]+vb
```

**cycle through all UVLs and determine reactions**

```python
UVL_record=np.empty([0,2])
if(nUVL)>0:
    for n,p in enumerate(linearloads):
        va,vb=reactions_UVL(n)
        UVL_record=np.append(UVL_record,[np.array([va,vb])],axis=0)
        
         #add reactions to the record
        reactions[0]=reactions[0]+va
        reactions[2]=reactions[2]+vb
```

# Shear and moment calculation

**define a function to calculate shear force and bending moment due to point load**

```python
def shear_moment_PL(n):
    xp=pointloads[n,0]
    fx=pointloads[n,1]
    fy=pointloads[n,2]
    va=PL_record[n,0]       #stores va of the nth pointload
    vb=PL_record[n,2]       #stores vb of the nth pointload
    
    shearforce=np.zeros(len(X))    #initializing an array to hold value of shear force for all x co-ordinatesis for this pointload
    moment=np.zeros(len(X))   #initializing an array to hold value of bending moment for all x coordinates for this pointload
    
    for i,x in enumerate(X):
        Shear=0      #initializing value of shear fro given x coordinate
        Moment=0      #iniitalizing value of moment for given x coordinate
        
        if x>A:
            Shear=Shear+va
            Moment=Moment+(va*(x-A))
            
        if x>B:
            Shear=Shear+vb
            Moment=Moment+(vb*(x-B))
            
        if x>xp:
            Shear=Shear+fy
            Moment=Moment+(fy*(x-xp))
        
        shearforce[i]=Shear
        moment[i]=Moment
    
    return shearforce,moment
    
    
```

**define a function to calculate shear and bending moment due to point moments**

```python
def shear_moment_PM(n):
    xm=pointmoments[n,0]
    m=pointmoments[n ,1]
    va=PM_record[n,0]       #stores va of the nth poinTmoment
    vb=PM_record[n,1]       #stores vb of the nth pointmoment
    
    shearforce=np.zeros(len(X)) #initializing an array to hold value of shear force for all x co-ordinates for this pointmoment
    moment=np.zeros(len(X))   #initializing an array to hold value of bending moment for all x coordinates for this pointmoment
    
    for i,x in enumerate(X):
        Shear=0      #initializing value of shear fro given x coordinate
        Moment=0      #iniitalizing value of moment for given x coordinate
        
        if x>A:
            Shear=Shear+va
            Moment=Moment+(va*(x-A))
            
        if x>B:
            Shear=Shear+vb
            Moment=Moment+(vb*(x-B))
        
        if x>xm:
            #there is no influence of pointmoment on shear only on the bending moment
            Moment=Moment+m 
        
        shearforce[i]=Shear
        moment[i]=Moment
    
    return shearforce,moment
        
    
    
    
    
```

**define a function to calculate shear and bending moment due to UDL**

```python
def shear_moment_UDL(n):
    x_start=distributedloads[n,0]
    x_end=distributedloads[n ,1]
    fy=distributedloads[n,2]
    va=UDL_record[n,0]       #stores va of the nth UDL
    vb=UDL_record[n,1]       #stores vb of the nth UDL
    
    shearforce=np.zeros(len(X)) #initializing an array to hold value of shear force for all x co-ordinates for this UDL
    moment=np.zeros(len(X))   #initializing an array to hold value of bending moment for all x coordinates for this UDL
    
    for i,x in enumerate(X):
        Shear=0      #initializing value of shear fro given x coordinate
        Moment=0      #iniitalizing value of moment for given x coordinate
        
        if x>A:
            Shear=Shear+va
            Moment=Moment+(va*(x-A))
            
        if x>B:
            Shear=Shear+vb
            Moment=Moment+(vb*(x-B))
        
        if(x>x_start and x<=x_end):
            Shear=Shear+(fy*(x-x_start))
            Moment=Moment+ fy*(x-x_start)*0.5*(x-x_start)
        elif(x>x_end):
            Shear=Shear+fy*(x_end-x_start)
            Moment=Moment+fy*(x_end-x_start)(x-(x_start+0.5*(x_end-x_start)))
        
        shearforce[i]=Shear
        moment[i]=Moment
    
    return shearforce,moment
```

**define a function to calculate shear and bending moment due to UVL**

```python
def shear_moment_UVL(n):
    x_start=linearloads[n,0]
    x_end=linearloads[n ,1]
    fy_start=linearloads[n,2]
    fy_end=linearloads[n,3]
    va=UVL_record[n,0]       #stores va of the nth UVL
    vb=UVL_record[n,1]       #stores vb of the nth UVl
    
    shearforce=np.zeros(len(X)) #initializing an array to hold value of shear force for all x co-ordinates for this UVL
    moment=np.zeros(len(X))   #initializing an array to hold value of bending moment for all x coordinates for this UVL
    
    for i,x in enumerate(X):
        Shear=0              #initializing value of shear fro given x coordinate
        Moment=0             #iniitalizing value of moment for given x coordinate
        
        if x>A:
            Shear=Shear+va
            Moment=Moment+(va*(x-A))
            
        if x>B:
            Shear=Shear+vb
            Moment=Moment+(vb*(x-B))
        
        if(fy_end==0):
            if(x>x_start and x<=x_end):
                    fy=((x_end-x)/(x_end-x_start))*fy_start
                    if(x>B):
                        Shear=-(0.5*fy*(x_end-x))
                        Moment=(0.5*fy*(x_end-x))*((x_end-x)/3)
                    if(x<B):
                        Shear=-(vb+(0.5*fy*(x_end-x)))
                        Moment=(vb*(B-x))+(0.5*fy*(x_end-x))*((x_end-x)/3)
            if(x>x_end):
                Shear=Shear+(0.5*fy_start*(x_end-x_start))
                Moment=Moment+(0.5*fy_start*(x_end-x_start))*(x-(x_start+((x_end-x_start)/3)))
        
        if(fy_start==0):
            if(x>x_start and x<=x_end):
                    fy=((x-x_start)/(x_end-x_start))*fy_end
                    Shear=Shear+(0.5*fy*(x-x_start))
                    Moment=Moment+(0.5*fy*(x-x_start))*((x-x_start)/3)
                    
            if(x>x_end):
                Shear=Shear+(0.5*fy_start*(x_end-x_start))
                Moment=Moment+(0.5*fy_start*(x_end-x_start))*(x-(x_start+((x_end-x_start)/(3/2))))
        
            
        shearforce[i]=Shear
        moment[i]=Moment
    
    return shearforce,moment
```

```python
if nUDL>0:
    for n,p in enumerate(distributedloads):
        shear,bending_moment=shear_moment_UDL(n)
        shear1=np.append(shear1,[shear],axis=0)                #stores the value of shearforce for a UDL
        moment=np.append(moment,[bending_moment],axis=0)       #stores the value of bm for a UDL
        
```

**cycle through all UVLs and determine shear and BM**

```python
if nUVL>0:
    for n,p in enumerate(linearloads):
        shear,bending_moment=shear_moment_UVL(n)
        shear1=np.append(shear1,[shear],axis=0)                #stores the value of shearforce for a UVL
        moment=np.append(moment,[bending_moment],axis=0)       #stores the value of bm for a UVL
        
```
